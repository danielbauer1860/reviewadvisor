```python
import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
ott_dataset = pd.read_csv("/content/drive/MyDrive/data/ott_et_al_processed.csv", sep=";")
```

```python
fake_review_dataset = pd.read_csv("/content/drive/MyDrive/data/final_review_dataset.csv", sep=";")
```

```python
fake_review_dataset.head()
```

```python
fr_train, fr_test = train_test_split(fake_review_dataset, test_size=0.2, shuffle=True, stratify=None, random_state=1860)
ott_train, ott_test = train_test_split(ott_dataset, test_size=0.2, shuffle=True, stratify=None, random_state=1860)
```

```python
fr_train.head()
```

```python
COMMENT = 'text'
fr_train[COMMENT].fillna("unknown", inplace=True)
fr_test[COMMENT].fillna("unknown", inplace=True)

ott_train[COMMENT].fillna("unknown", inplace=True)
ott_test[COMMENT].fillna("unknown", inplace=True)
```

# Building the model

```python
import re, string
re_tok = re.compile(f'([{string.punctuation}“”¨«»®´·º½¾¿¡§£₤‘’])')
def tokenize(s): return re_tok.sub(r' \1 ', s).split()
```

```python
n = fr_train.shape[0]
vec = TfidfVectorizer(ngram_range=(1,2), tokenizer=tokenize,
               min_df=3, max_df=0.9, strip_accents='unicode', use_idf=1,
               smooth_idf=1, sublinear_tf=1 )
fr_trn_term_doc = vec.fit_transform(fr_train[COMMENT])
fr_test_term_doc = vec.transform(fr_test[COMMENT])

ott_trn_term_doc = vec.fit_transform(ott_train[COMMENT])
ott_test_term_doc = vec.transform(ott_test[COMMENT])
```
